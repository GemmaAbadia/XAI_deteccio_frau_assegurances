## Neteja del dataset original

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importar dataset original
df = pd.read_csv("insurance_claims.csv")

# Eliminar columnes inútils
cols_to_drop = ["_c39", "policy_number", "incident_location"]  # IDs i columnes sense info útil
df_clean = df.drop(columns=cols_to_drop)

# Convertir dates a datetime
df_clean["policy_bind_date"] = pd.to_datetime(df_clean["policy_bind_date"])
df_clean["incident_date"] = pd.to_datetime(df_clean["incident_date"])

# Mostrar les primeres files per comprovar
df_clean.head()
```

```python
# Crear columna binària per la variable de frau
df_clean["fraud_binary"] = df_clean["fraud_reported"].map({"N": 0, "Y": 1})

# Ja no necessitem la columna original de "fraud_reported" si vols
df_clean = df_clean.drop(columns=["fraud_reported"])

# Mostrar les primeres files per comprovar
df_clean.head()
```

```python
# Guardar el dataset net
df_clean.to_csv("insurance_claims_clean.csv", index=False)
```
