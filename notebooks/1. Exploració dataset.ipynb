## Comprovació de model base

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_auc_score

# Copia del dataframe
df_model = df.copy()

# Convertir target a binari si cal
le = LabelEncoder()
df_model['fraud_reported'] = le.fit_transform(df_model['fraud_reported'])

# Convertir categòriques
df_model = pd.get_dummies(df_model, drop_first=True)

X = df_model.drop("fraud_reported", axis=1)
y = df_model["fraud_reported"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict_proba(X_test)[:,1]

roc_auc_score(y_test, y_pred)
```

```python
# Guardem en un CSV nou
df_model.to_csv("insurance_claims_clean.csv", index=False)
```
